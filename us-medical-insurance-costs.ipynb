# U.S. Medical Insurance Costs

This project is going to analyse an **Insurance.CSV** file containing **U.S. Medical Insurance Records** using **Python Fundamentals**. The goal of this project will be to answer the following questions:

1. What is the average age?
2. What is the youngest and oldest age?
3. What is the average insurance cost?
5. Who paid more for insurance, Male or Female?
6. Who paid more for insurance, smoker or non-smoker?
7. Who is healthier, Male or female?
8. Which region has the most children?

```python
# import CSV library which gives the tools to parse the CSV file
import csv

# read the CSV file
with open('insurance.csv', 'r') as insurance_csv_file:
    insurance_csv_dict = csv.DictReader(insurance_csv_file)
    insurance_dict = {'age': [], 'sex': [], 'bmi': [], 'children': [], 'smoker': [], 'region': [], 'charges': [] }
    for row in insurance_csv_dict:
        insurance_dict['age'].append(row['age'])
        insurance_dict['sex'].append(row['sex'])
        insurance_dict['bmi'].append(row['bmi'])
        insurance_dict['children'].append(row['children'])
        insurance_dict['smoker'].append(row['smoker'])
        insurance_dict['region'].append(row['region'])
        insurance_dict['charges'].append(row['charges'])
#print(insurance_dict)
        
```

```python
# What is the average age?
total_age = 0
for age in insurance_dict['age']:
    total_age += int(age)
print('Average age: ' + str(round(total_age/len(insurance_dict['age']))) + ' years old.')
```

```python
# How old are the youngest patients?

for youngest_age in insurance_dict['age']:
    print(min(insurance_dict['age']))
```

```python
# How old are the oldest patients?

for age in insurance_dict['age']:
    print(max(insurance_dict['age']))

```

```python
# What is the average insurance cost?
total_insurance_cost = 0
for cost in insurance_dict['charges']:
    total_insurance_cost += float(cost)
print('Average Insurance cost: ' + str(round(total_insurance_cost/len(insurance_dict['charges']))) + ' dollars.')
```

```python
# Who paid more for insurance, Male or Female?

def insurance_cost(cost):
    male_patients = [charge for sex, charge in zip(cost['sex'], cost['charges']) if sex == 'male']
    female_patients = [charge for sex, charge in zip(cost['sex'], cost['charges']) if sex == 'female']
    male_total = 0 
    female_total = 0 
    for male, female in zip(male_patients, female_patients):
        male_total += float(male)
        female_total += float(female)
    print('Average Male Insurance Cost: ' + str(round(male_total/len(male_patients))) + ' dollars.')
    print('Average Female Insurance Cost: ' + str(round(female_total/len(female_patients))) + ' dollars.')
    
    male_avg = male_total/len(male_patients)
    female_avg = female_total/len(female_patients)
    difference = male_avg - female_avg
    print('The Male paid ' + str(round(difference, 2)) + ' dollars more than the Female. ')
insurance_cost(insurance_dict)
```

```python
# Who paid more for insurance, smoker or non-smoker?

def smoker(cost):
    smoker_patients = [charge for smoker, charge in zip(cost['smoker'], cost['charges']) if smoker == 'yes']
    non_smoker_patients = [charge for smoker, charge in zip(cost['smoker'], cost['charges']) if smoker == 'no']
    smoker_total = 0 
    non_smoker_total = 0 
    for smoker, non_smoker in zip(smoker_patients, non_smoker_patients):
        smoker_total += float(smoker)
        non_smoker_total += float(non_smoker) 
    print('Average Smoker Insurance Cost: ' + str(round(smoker_total/len(smoker_patients), 2)) + ' dollars.')
    print('Average Non_Smoker Insurance Cost: ' + str(round(non_smoker_total/len(non_smoker_patients), 2)) + ' dollars.')
    
    smoker_avg = smoker_total/len(smoker_patients)
    non_smoker_avg = non_smoker_total/len(non_smoker_patients)
    difference = smoker_avg - non_smoker_avg
    print('The Smokers paid ' + str(round(difference, 2)) + ' dollars more than the Non-Smokers.')
smoker(insurance_dict)
```

```python
# Who is healthier, Male or female?

def bmi(cost):
    male_patients = [bmi for sex, bmi in zip(cost['sex'], cost['bmi']) if sex == 'male']
    female_patients = [bmi for sex, bmi in zip(cost['sex'], cost['bmi']) if sex == 'female']
    male_total = 0
    female_total = 0
    for male, female in zip(male_patients, female_patients):
        male_total += float(male)
        female_total += float(female)
    print('Average Male BMI: ' + str(round(male_total/len(male_patients), 2)))
    print('Average Female BMI: ' + str(round(female_total/len(female_patients), 2)))
bmi(insurance_dict)
```

```python
# Which region has the most children?

def region(cost):
    south_west = [child for region, child in zip(cost['region'], cost['children']) if region == 'southwest']
    south_east = [child for region, child in zip(cost['region'], cost['children']) if region == 'southeast']
    north_west = [child for region, child in zip(cost['region'], cost['children']) if region == 'northwest']
    north_east = [child for region, child in zip(cost['region'], cost['children']) if region == 'northeast']
    south_west_total = 0
    south_east_total = 0
    north_west_total = 0
    north_east_total = 0
    for south_w, south_e in zip(south_west, south_east):
        south_west_total += int(south_w)
        south_east_total += int(south_e)
    for north_w, north_e in zip(north_west, north_east):
        north_west_total += int(north_w)
        north_east_total += int(north_e)
    print('Total Number of Children in South West: ' + str(south_west_total))
    print('Total Number of Children in South East: ' + str(south_east_total))
    print('Total Number of Children in North West: ' + str(north_west_total))
    print('Total Number of Children in North East:' + str(north_east_total))
    
region(insurance_dict)
```
